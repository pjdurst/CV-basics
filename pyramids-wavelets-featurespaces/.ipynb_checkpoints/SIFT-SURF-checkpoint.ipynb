```python
# Starting with the usual imports
import numpy as np
import matplotlib as mp
from matplotlib import pyplot as plt
import cv2
from PIL import Image
import time
```

```python
# Basic, out-of-the box SIFT with openCV
# Load the image
image = cv2.imread("RIT-1.jpg", cv2.IMREAD_GRAYSCALE)

# Initialize the SIFT detector
sift = cv2.SIFT_create()

# Detect keypoints and compute descriptors
keypoints, descriptors = sift.detectAndCompute(image, None) # 'None' refers to a Mask parameter that simply tells SIFT to only detect keypoints in a certain region of the image

# Draw keypoints on the image
sift_image = cv2.drawKeypoints(image, keypoints, None, flags=cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)

# Display the image with keypoints
plt.figure(figsize=(10, 6))
plt.imshow(sift_image, cmap="gray")
plt.axis("off")
plt.title("SIFT Keypoints")
plt.show()
```

The basic SIFT implemenation is quite noisy, and we often need to use the Mask parameter. In genreal, just pointing SIFT at an image isn't helpful. If we don't provide a source image, SIFT pulls out everything that looks like a keypoint, e.g., everywhere that has a decent intensity change in one direction. So, do we hate ourselves enough to write our own SIFT implementation?

```python
# The first step - apply Gaussian blur
def gaussian_blur(image, sigma):
    """Apply Gaussian blur using a given sigma."""
    return scipy.ndimage.gaussian_filter(image, sigma)

# The second step - function to create our scale space
# Number of octaves = the number of scale space layers to create
# Number of scales - the number of images to create in each scale space
# So, we have a four-layer pyramid with 5 scaled images in each layer
def generate_gaussian_pyramid(image, num_octaves=4, num_scales=5, sigma=1.6):
    """Generate a Gaussian pyramid with different scales."""
    pyramid = [] #Initialize the image pyramid / scale space
    k = 2 ** (1 / num_scales)  # Arbitrary scale factor
    for _ in range(num_octaves):
        octave = [gaussian_blur(image, sigma * (k ** i)) for i in range(num_scales)] # Craete the octave
        pyramid.append(octave) # Add the octave to the pyramid
        image = octave[-1][::2, ::2]  # Downsample the image for next octave
    return pyramid

# The third step - function to create of Difference of Gaussians
# In SIFT, we are looking at the Difference of Gaussians --between-- image octaves; that is, we want the DoG of the difference of the images, not the difference of the Gaussians within a single image
# For SIFT, we need to apply Gaussian blur to each image in each octave
# For programming flow and data structure consistency, we will create another image pyramid made up of differences between the blurred images at each octave
def compute_dog_pyramid(gaussian_pyramid):
    """Compute Difference of Gaussian (DoG) pyramid."""
    dog_pyramid = []
    for octave in gaussian_pyramid:
        dog_pyramid.append([octave[i+1] - octave[i] for i in range(len(octave) - 1)]) # Find the DoG between octaves
    return dog_pyramid

# At this point, we are looking at the differences in the difference between image scales.
# Put differently, applying a DoG to an image finds areas of change in that image. Applying a DoG to an image pyramid finds of areas of change between two images at different scales


# Find the keypoints in the image
def detect_keypoints(dog_pyramid):
    """Find keypoints as extrema in DoG pyramid."""
    keypoints = []
    for octave_idx, octave in enumerate(dog_pyramid): # for each DoG image in the DoG pyramid
        for scale_idx in range(1, len(octave) - 1): # for each scale in each octave of the DoG pyramid
            img_prev, img_curr, img_next = octave[scale_idx - 1], octave[scale_idx], octave[scale_idx + 1] # Look for the similarities between images
            for i in range(1, img_curr.shape[0] - 1):
                for j in range(1, img_curr.shape[1] - 1): # check each pixel
                    pixel = img_curr[i, j]
                    neighbors = np.hstack([
                        img_curr[i-1:i+2, j-1:j+2].flatten(),
                        img_prev[i-1:i+2, j-1:j+2].flatten(),
                        img_next[i-1:i+2, j-1:j+2].flatten()
                    ])
                    if pixel == np.max(neighbors) or pixel == np.min(neighbors): # if the pixel is within the threshold, mark it as a keypoint
                        keypoints.append((octave_idx, scale_idx, i, j))
    return keypoints


# Plot the keypoints
def visualize_keypoints(image, keypoints):
    """Visualize keypoints on the image."""
    plt.figure(figsize=(8, 6))
    plt.imshow(image, cmap='gray')
    for _, _, y, x in keypoints:
        plt.scatter(x, y, c='red', s=5)
    plt.title("SIFT Keypoints")
    plt.axis("off")
    plt.show()

# Load the initial image as grayscale
image = cv2.imread("RIT-1.jpg", cv2.IMREAD_GRAYSCALE)
image = image.astype(np.float32) / 255.0  # Normalize, since we don't have openCV handling this for us

# Generate Gaussian and DoG pyramids
gaussian_pyramid = generate_gaussian_pyramid(image)
dog_pyramid = compute_dog_pyramid(gaussian_pyramid)

# Detect keypoints
keypoints = detect_keypoints(dog_pyramid)

# Visualize detected keypoints
visualize_keypoints(image, keypoints)
```

Totally worth it...

```python
# Now let's try a SURF implementation
# Load the image
image = cv2.imread("RIT-1.jpg", cv2.IMREAD_GRAYSCALE)

# Create a SURF detector object (Set Hessian threshold)
surf = cv2.xfeatures2d.SURF_create(hessianThreshold=400)
# Detect keypoints and descriptors
keypoints, descriptors = surf.detectAndCompute(image, None)

# Draw keypoints on the image
image_with_keypoints = cv2.drawKeypoints(image, keypoints, None, (255, 0, 0), 4)

# Display the image
plt.figure(figsize=(10, 6))
plt.imshow(image_with_keypoints, cmap='gray')
plt.title("SURF Keypoints")
plt.axis("off")
plt.show()
```

SURF is not a free algorithm, because it's hard and complicated and, more than anything, vital to a lot of CV, so why do it for free? Instead, let's look at the ORB (Oriented FAST and Rotated BRIEF) Algorithm

```python
# Load the image
image = cv2.imread("RIT-1.jpg", cv2.IMREAD_GRAYSCALE)

# Create ORB detector
orb = cv2.ORB_create()

# Detect keypoints and compute descriptors
keypoints, descriptors = orb.detectAndCompute(image, None)

# Draw keypoints on the image
image_with_keypoints = cv2.drawKeypoints(image, keypoints, None, color=(0, 255, 0), flags=0)

# Display the image
plt.figure(figsize=(10, 6))
plt.imshow(image_with_keypoints, cmap='gray')
plt.title("ORB Keypoints")
plt.axis("off")
plt.show()
```

Right away, we see this both ran MUCH faster than our hand-built SIFT but also provides way more accurate / less noisy results. Let's look at the ORB parameters

```python
orb = cv2.ORB_create(
    nfeatures=500,       # Number of keypoints to detect (default: 500)
    scaleFactor=1.2,     # Pyramid scale factor (default: 1.2)
    nlevels=8,           # Number of pyramid levels (default: 8)
    edgeThreshold=31,    # Edge threshold (default: 31)
    patchSize=31         # Patch size (default: 31)
)
```

```python
# Load images
image1 = cv2.imread("RIT.jpg", cv2.IMREAD_GRAYSCALE)
image2 = cv2.imread("RIT2.jpg", cv2.IMREAD_GRAYSCALE)

# Initialize SIFT detector
sift = cv2.SIFT_create()

# Initialize ORB detector
orb = cv2.ORB_create(nfeatures=500)

# Detect and compute keypoints and descriptors with SIFT
kp1_sift, des1_sift = sift.detectAndCompute(image1, None)
kp2_sift, des2_sift = sift.detectAndCompute(image2, None)

# Detect and compute keypoints and descriptors with ORB
kp1_orb, des1_orb = orb.detectAndCompute(image1, None)
kp2_orb, des2_orb = orb.detectAndCompute(image2, None)

# Using FLANN-based matcher for SIFT
# FLANN parameters for SIFT
index_params = dict(algorithm=1, trees=5)  # KD-tree
search_params = dict(checks=50)           # Recursions

flann = cv2.FlannBasedMatcher(index_params, search_params)
matches_sift = flann.knnMatch(des1_sift, des2_sift, k=2)

# Apply ratio test to keep good matches in SIFT
good_matches_sift = []
for m, n in matches_sift:
    if m.distance < 0.7 * n.distance:
        good_matches_sift.append(m)

# Brute-Force Matcher for ORB (since descriptors are binary)
bf = cv2.BFMatcher(cv2.NORM_HAMMING, crossCheck=True)
matches_orb = bf.match(des1_orb, des2_orb)

# Sort matches by distance for ORB (optional)
matches_orb = sorted(matches_orb, key=lambda x: x.distance)

# Draw matches for both SIFT and ORB
sift_img_matches = cv2.drawMatches(image1, kp1_sift, image2, kp2_sift, good_matches_sift, None, flags=cv2.DrawMatchesFlags_NOT_DRAW_SINGLE_POINTS)
orb_img_matches = cv2.drawMatches(image1, kp1_orb, image2, kp2_orb, matches_orb[:50], None, flags=cv2.DrawMatchesFlags_NOT_DRAW_SINGLE_POINTS)

# Display the results side-by-side
plt.figure(figsize=(20,10))

plt.subplot(1, 2, 1)
plt.title("SIFT Matches")
plt.imshow(sift_img_matches)

plt.subplot(1, 2, 2)
plt.title("ORB Matches")
plt.imshow(orb_img_matches)

plt.show()
```

```python
import cv2

# Load images in grayscale
image1 = cv2.imread("RIT.jpg", cv2.IMREAD_GRAYSCALE)
image2 = cv2.imread("RIT2.jpg", cv2.IMREAD_GRAYSCALE)

# Initialize SIFT or SURF detector
sift = cv2.SIFT_create()
# surf = cv2.xfeatures2d.SURF_create(hessianThreshold=400) # Uncomment if using SURF

# Detect and compute keypoints and descriptors
kp1, des1 = sift.detectAndCompute(image1, None)
kp2, des2 = sift.detectAndCompute(image2, None)

# Using FLANN-based matcher
index_params = dict(algorithm=1, trees=5) # FLANN index for KD-tree
search_params = dict(checks=50)           # Number of times the tree is recursively traversed

flann = cv2.FlannBasedMatcher(index_params, search_params)
matches = flann.knnMatch(des1, des2, k=2)

# Apply ratio test
good_matches = []
for m, n in matches:
    if m.distance < 0.7 * n.distance:
        good_matches.append(m)

# Draw matches
result = cv2.drawMatches(image1, kp1, image2, kp2, good_matches, None, flags=cv2.DrawMatchesFlags_NOT_DRAW_SINGLE_POINTS)

# Display matches
cv2.imshow("Matches", result)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
