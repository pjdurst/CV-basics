Let's look into the HSV color space
We'll start by creating our own HSV transformer

```python
# Starting with the usual imports
import numpy as np
import matplotlib as mp
from matplotlib import pyplot as plt
import cv2
from PIL import Image
```

```python
# Load the RIT image
baseImg = cv2.imread('godzilla.jpg')
rgb = cv2.cvtColor(baseImg, cv2.COLOR_BGR2RGB)
plt.imshow(rgb, cmap = plt.cm.Spectral)
plt.show()
```

```python
# Let openCV convert for us
baseImg = cv2.imread('godzilla.jpg')
img = cv2.cvtColor(baseImg, cv2.COLOR_BGR2HSV)
plt.imshow(img, cmap = plt.cm.Spectral)
plt.show()
```

```python
# Can we do this manually?
# Load image
rgb = cv2.imread('godzilla.jpg')
# Normalize RGB values to [0, 1]
rgb_image = rgb / 255.0
    
# Initialize the HSV image
hsv_image = np.zeros_like(rgb_image)
    
# Extract individual channels
R = rgb_image[..., 0]
G = rgb_image[..., 1]
B = rgb_image[..., 2]
    
# Compute Cmax, Cmin, and Delta
Cmax = np.max(rgb_image, axis=-1)
Cmin = np.min(rgb_image, axis=-1)
Delta = Cmax - Cmin
    
# Hue calculation
H = np.zeros_like(Cmax)
H[Cmax == R] = (60 * ((G[Cmax == R] - B[Cmax == R]) / Delta[Cmax == R])) + 360
H[Cmax == G] = (60 * ((B[Cmax == G] - R[Cmax == G]) / Delta[Cmax == G])) + 120
H[Cmax == B] = (60 * ((R[Cmax == B] - G[Cmax == B]) / Delta[Cmax == B])) + 240
H = (H + 360) % 360  # Ensure Hue is in range [0, 360]
    
# Saturation calculation
S = np.zeros_like(Cmax)
S[Cmax != 0] = Delta[Cmax != 0] / Cmax[Cmax != 0]
    
# Value calculation
V = Cmax
    
# Create the HSV image array
hsv_image[..., 0] = H / 360.0  # Hue in range [0, 1]
hsv_image[..., 1] = S          # Saturation in range [0, 1]
hsv_image[..., 2] = V          # Value in range [0, 1]
    
# If you want to view the result, you can use OpenCV to show the image
# Convert to 0-255 range for display (optional)
hsv_image_display = (hsv_image * 255).astype(np.uint8)
cv2.imshow('HSV Image', hsv_image_display)

# Plot the original and processed image
plt.figure(figsize=(12, 6))

# Original Image
plt.subplot(1, 2, 1)
plt.title('Original Image')
plt.imshow(rgb)
plt.axis('off')

# Processed Image (in RGB)
plt.subplot(1, 2, 2)
plt.title('Processed Image (RGB to HSV)')
plt.imshow(hsv_image)
plt.axis('off')
```

Looks like we forgot to change our initial images r,g,b coordinate space!

This is why paying attention to pixel coordinates is so important.
